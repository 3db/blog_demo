```python
import pandas as pd
import numpy as np
import json
```

```python
class_map = json.load(open('../output/part_of_object/class_maps.json'))
df = pd.DataFrame.from_records(list(map(json.loads, open('../output/part_of_object/details.log').readlines())))
df = df.drop('render_args', axis=1).join(pd.DataFrame(df.render_args.values.tolist()))
df['prediction'] = df['prediction'].apply(lambda x: class_map[x[0]])
df['is_correct'] = (df['is_correct'] == 'True')
df['is_correct'] = df['prediction'].isin(['cup', 'coffee mug'])
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from PIL import Image
from tqdm import tqdm
```

```python
uv_corrects = np.zeros((256, 256))
uv_tots = np.zeros((256, 256))
for _id in tqdm(df["id"].unique().tolist()):
    uv_coords = np.array(Image.open(f'../output/part_of_object/images/{_id}_uv.png'))[:,:,:2]
    uv_coords = uv_coords.reshape(-1, 2)
    uv_tots[uv_coords[:,0],uv_coords[:,1]] += 1.
    uv_corrects[uv_coords[:,0],uv_coords[:,1]] += float(df.set_index('id').loc[_id]['is_correct'])
```

```python
Image.fromarray((255 * (uv_corrects / (uv_tots + 1e-4))).astype('uint8'))
```
